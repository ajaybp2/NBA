```python
df_draft1 = df[df['draft_round'] == '1']
sns.displot(df_draft1['pts'], kde=True)
plt.xlabel('Points Scored')
plt.ylabel('Number of Players')
plt.title('Points Scored By Draft 1 Picks')

df_draft1 = df[df['draft_round'] == '1']
sns.displot(df_draft1['ast'], kde=True)
plt.xlabel('Assists')
plt.ylabel('Number of Players')
plt.title('Assists By Draft 1 Picks')

df_draft1 = df[df['draft_round'] == '1']
sns.displot(df_draft1['reb'], kde=True)
plt.xlabel('Rebounds')
plt.ylabel('Number of Players')
plt.title('Rebounds By Draft 1 Picks')
```

```python
df_draft2 = df[df['draft_round'] == '2']
sns.displot(df_draft2['pts'], kde=True)
plt.xlabel('Points Scored')
plt.ylabel('Number of Players')
plt.title('Points Scored By Draft 2 Picks')

df_draft2 = df[df['draft_round'] == '2']
sns.displot(df_draft2['ast'], kde=True)
plt.xlabel('Assists')
plt.ylabel('Number of Players')
plt.title('Assists By Draft 2 Picks')

df_draft2 = df[df['draft_round'] == '2']
sns.displot(df_draft2['reb'], kde=True)
plt.xlabel('Rebounds')
plt.ylabel('Number of Players')
plt.title('Rebounds By Draft 2 Picks')
```


    From the data analysis I was able to answer the questions proposed at the beginning of the project. I have found that height does play a major role in the amount of rebounds a NBA player gets. The average height for an NBA player is 200.8 cm. Players under that height on avearge grabbed 2.6 rebounds and players over that average grabbed 4.6 rebounds. This is an 80% increase in amounts of rebounds grabbed. The height where the most rebounds were grabbed is around 210 cm. 
    
    I was also able to find that players chosen in the 1sr round of the draft are on average become better players than those chosen in the 2nd round of the draft. Players that were chosen in the first round averaged 10 ppg, 2.1 apg, and 4.2 rpg. Whereas players chosen in the second draft average 6.2 ppg, 1.3 apg, and 2.9 rbp. Players in the first draft are better in every category by a reasonable amount.
